### Wrangle

```python
#acquire
colnames=['ip', 'timestamp', 'request_method', 'status', 'size', 'destination', 'request_agent']

df_orig = pd.read_csv('http://python.zach.lol/access.log', 
                      engine='python', 
                      header=None, 
                      index_col=False, 
                      names=colnames, 
                      sep=r'\s(?=(?:[^"]*"[^"]*")*[^"]*$)(?![^\[]*\])', 
                      na_values='"-"', 
                      usecols=[0, 3, 4, 5, 6, 7, 8])

new = pd.DataFrame([["95.31.18.119", "[21/Apr/2019:10:02:41+0000]", 
                     "GET /api/v1/items/HTTP/1.1", 200, 1153005, np.nan, 
                     "python-requests/2.21.0"],
                    ["95.31.16.121", "[17/Apr/2019:19:36:41+0000]", 
                     "GET /api/v1/sales?page=79/HTTP/1.1", 301, 1005, np.nan, 
                     "python-requests/2.21.0"],
                    ["97.105.15.120", "[18/Apr/2019:19:42:41+0000]", 
                     "GET /api/v1/sales?page=79/HTTP/1.1", 301, 2560, np.nan, 
                     "python-requests/2.21.0"],
                    ["97.105.19.58", "[19/Apr/2019:19:42:41+0000]", 
                     "GET /api/v1/sales?page=79/HTTP/1.1", 200, 2056327, np.nan, 
                     "python-requests/2.21.0"]], columns=colnames)

df = df_orig.append(new)
df.head()
```

```python
#communication log files
df.info()
```

```python
#prepare: timestamp values need to be cleaned by removing the brackets and extraneous tail
df.timestamp = df.timestamp.str.replace(r'(\[|\+0000\])', '', regex=True)
df.head(2)
```

```python
#replace the first colon in timestamp before year to accomodate for pandas datetime
df.timestamp = pd.to_datetime(df.timestamp.str.replace(':', ' ', 1))

#reset index to timestamp
df = df.set_index('timestamp')

df.head(2)
```

```python
#clean up strings in columns with extra quotes (remove them/replace them w/ nothing)
for col in ['request_method', 'request_agent', 'destination']: 
    df[col] = df[col].str.replace('"', '')
    
#shorten values in request method and clean up by removing page data
df['request_method'] = df.request_method.str.replace(r'\?page=[0-9]+', '', regex=True)

df.head()
```

```python
#add variable - converting size of transmission (bytes to megabytes (mb)):
df['size_mb'] = [n/1024/1024 for n in df['size']]
df.head()
```

```python
df.describe()
```

### Detecting Anomalies in Discrete Variables:
#### Finding anomalies in already existing data

```python
#get the value counts of request_method in the dataset
pd.DataFrame(df.request_method.value_counts(dropna=False)).reset_index()

#col names need to be renamed (next cell)
```

```python
#resetting index makes it a dataframe
reqm_df = df.request_method.value_counts(dropna=False).reset_index()\
                .rename(index=str, columns={'index': 'request_method', 'request_method': 'rm_count'})

reqm_df.head()
```

```python
#get the probability of request_method in the dataset by dividing value counts by count
reqm_df2 = (df.request_method.value_counts(dropna=False) / df.request_method.count()).reset_index().\
                rename(columns={'index': 'request_method', 'request_method': 'reqm_proba'})

reqm_df2.head()
```

```python
#merge/join the two dfs together (reqm_df and reqm_df2)
reqm_df = reqm_df.merge(reqm_df2)
reqm_df.head()
```

```python
#compare shape of original df
df.shape, reqm_df.shape
```

```python
#alter the reqm_df to see only those where rate < 1% 
reqm_df[reqm_df.reqm_proba < .01]
reqm_df
```

```python
#look for anomalies 
print(len(reqm_df))

print(reqm_df.tail(10))

plt.figure(figsize=(15, 8))
splot = sns.barplot(data=reqm_df, x = 'request_method', y = 'rm_count', ci = None)

#annotate bars based on its height
for p in splot.patches:
    splot.annotate(format(p.get_height(), '.0f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', xytext = (0, 10), 
                   textcoords = 'offset points')
plt.xticks(rotation='vertical')
plt.show()
```

### Another way: Detecting anomalies by establishing a baseline and evaluate as new data arrives.

```python
#establish baseline and create train set
train = df.loc['2019-04-16 19:34:42':'2019-04-17 12:55:14']

#set train columns
train = train[['ip','request_method','status','size','destination','request_agent','size_mb']]
train.head()
```

```python
#check train size
train.size/df.size
```

```python
#compute probabilities based on train sample
reqm_df = pd.DataFrame(train.request_method.value_counts(dropna=False) / train.request_method.count()).reset_index().\
                rename(index=str, columns={'index': 'request_method', 'request_method': 'reqm_proba'})

reqm_df
```

```python
#merge proabilities with all data (train + new data): 
#where the request_method is new, (not seen in training set), fill the prob with a value of zero

df = df.reset_index().merge(reqm_df, on=['request_method'], how='left').fillna(value=0).set_index('timestamp')
df.reqm_proba.value_counts()

df
```

### Conditional Probabilities: Probabilities using 2 discrete variables

```python
#prob (A|B) = prob(A & B) / prob(B)

#using _future_ division (.div)
reqm_probs = train.groupby('request_method').size().div(len(df))

#status_given_rm is prob(A) 
#train.groupby is getting the prob(A&B) / then dividing it by just the request_method probability
#making sure we are dividing on the request_method level
status_given_reqm = pd.DataFrame(train.groupby(['request_method', 'status']).\
                               size().div(len(train)).\
                               div(reqm_probs, 
                                   axis=0, 
                                   level='request_method').\
                               reset_index().\
                               rename(index=str, 
                                      columns={0: 'proba_status_given_reqm'}))

status_given_reqm
```

```python
reqm_status_count = pd.DataFrame(train.groupby(['request_method', 'status'])['size'].\
                                count().reset_index().\
                                rename(index=str, 
                                       columns={'size': 'reqm_status_count'}))


reqm_status = status_given_reqm.merge(reqm_status_count)
reqm_status
```

```python
#add these probabilities to original events to detect anomalous events:
#filling w/ 0 prob anything we haven't seen before
df = df.reset_index().merge(reqm_status, on=['request_method', 'status'], how='left').fillna(value=0).set_index('timestamp')
df.head()
```

```python
#visualize w/ scatter plot
plt.scatter(df.proba_status_given_reqm, df.reqm_proba)
```

```python
df[(df['proba_status_given_reqm'] < 0.01) & (df['reqm_proba'] >0.8)]
```

```python
#look for anomalies 

plt.figure(figsize=(15, 8))
splot = sns.barplot(data=reqm_df, x = 'request_method', y = 'reqm_proba', ci = None)

#annotate bars
for p in splot.patches:
    splot.annotate(format(p.get_height(), '.0f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', xytext = (0, 10), 
                   textcoords = 'offset points'
                   )
plt.xticks(rotation='vertical')
plt.show()
```

### Identifying anomalous request agents

```python
#get counts of request agents in dataset
ra_df = pd.DataFrame(df.request_agent.value_counts(dropna=False)).reset_index().\
                rename(index=str, columns={'index': 'request_agent', 'request_agent': 'request_agent_count'})

#get probability of request agents in the dataset
ra_df2 = pd.DataFrame(df.request_agent.value_counts(dropna=False)/df.request_agent.count()).reset_index().\
                rename(index=str, columns={'index': 'request_agent', 'request_agent': 'request_agent_proba'})

#merge these together
ra_df = ra_df.merge(ra_df2)

ra_df
```

```python
#see those where rate < 1% 
ra_df[ra_df.request_agent_proba < .01]
```

```python
#look for anomalies
print(len(ra_df))

print(ra_df.tail(10))

plt.figure(figsize=(15, 8))
splot = sns.barplot(data=ra_df, x = 'request_agent', y = 'request_agent_count', ci = None)
for p in splot.patches:
    #annotate
    splot.annotate(format(p.get_height(), '.0f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', xytext = (0, 10), 
                   textcoords = 'offset points'
                   )
    plt.xticks(rotation='vertical')
```
